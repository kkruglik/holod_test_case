Импортируем нужные библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
```

### Задание 1
Проанализируйте Gender pay gap в этих странах, сравните со средним результатом по EU.

Сравнивать будем Польшу и Испанию -- близкие по населению и по ВВП (ППС) страны

Загрузим датасет:

`gender_pay_gap` -- гендерный разрыв в оплате труда

```python
df = pd.read_csv('data/earn_gr_gpgr2_1_Data.csv')
df.head()
```

```python
df.info()
```

Помимо признаков страна, год и гендерный разрыв в процентах, в датафрейме также есть признак `NACE_R2`с указанием рабочей отрасли. Я буду анализировать агрегированную отрасль `Industry, construction and services (except activities of households as employers and extra-territorial organisations and bodies)`

```python
print(np.unique(df.NACE_R2))
```

Значения в столбце `Value` не в формате `int`. Предобработаем датафрейм

```python
df.Value = df.Value.apply(lambda x: float(x) if x != ':' else np.nan)

# уберём ненужные признаки
df.drop(['UNIT', 'Flag and Footnotes'], axis=1, inplace=True)
```

Создадим новый датафрейм с нужными странами и выбранной отраслью.

```python
gender_pay_gap = df[
    ((df.GEO == 'Poland') | (df.GEO == 'Spain')) & \
    (df.NACE_R2 == 'Industry, construction and services (except activities of households as employers and extra-territorial organisations and bodies)')
]
```

Посчитаем средний `Gender pay gap` для всех стран Евросоюза, и объединим датафремы в один

```python
gender_pay_gap_EU = df[df.NACE_R2 == 'Industry, construction and services (except activities of households as employers and extra-territorial organisations and bodies)']\
    .groupby('TIME').mean().reset_index()

gender_pay_gap_EU['NACE_R2'] = 'Industry, construction and services (except activities of households as employers and extra-territorial organisations and bodies)'
gender_pay_gap_EU['GEO'] = 'EU'

gender_pay_gap = pd.concat([gender_pay_gap, gender_pay_gap_EU], ignore_index=True)
```

```python
plt.subplots(figsize =(14, 8))
ax = sns.barplot(x='TIME', y='Value', hue='GEO', data=gender_pay_gap)

# for container in ax.containers:
#     ax.bar_label(container)
    
plt.title('Gender pay gap in Poland and Spain', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Gender pay gap (%)', fontsize=15)
plt.show()
```

**Выводы:**
- гендерное неравенство в зарплатах в Польше ниже, чем Испании и в среднем по странам Европы
- с 2010 по 2020 уровень гендерного неравенства в зарплатах в Польше рос, но к 2020 году вернулся к уровню 2010 года
- в Испании в начале 2010-х уровень гендерного неравенства в зарплатах был таким же, как в среднем по Европе. Но к началу 2020-х значительно сократился.

### Задание 2

Создайте отчет по датабазе `Labour market slack by sex and age — annual data` — сравните мужчин и женщин. Визуализируйте и прокомментируйте результаты.

```python
df = pd.read_csv('data/lfsi_sla_a_1_Data.csv')
```

```python
df.head(5)
```

Предобработаем датафрейм

```python
df.drop(['WSTATUS', 'UNIT', 'Flag and Footnotes', 'AGE'], axis=1, inplace=True)
```

```python
df.Value = df.Value.apply(lambda x: float(x) if x != ':' else np.nan)
```

```python
labour_slack = df[
    (df.GEO == 'Poland') | (df.GEO == 'Spain')
                               ]
```

Визуализируем данные

```python
plt.subplots(figsize =(14, 8))
ax = sns.barplot(x='TIME', y='Value', hue='SEX', data=labour_slack[labour_slack.GEO == 'Poland']) # palette='tab10', 
plt.title('Labor slack by sex in Poland', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Labor slack (%)', fontsize=15)

for container in ax.containers:
    ax.bar_label(container)
    
plt.show()
```

```python
plt.subplots(figsize =(14, 8))
ax = sns.barplot(x='TIME', y='Value', hue='SEX', data=labour_slack[labour_slack.GEO == 'Spain']) # palette='tab10', 
plt.title('Labor slack by sex in Spain', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Labor slack (%)', fontsize=15)

for container in ax.containers:
    ax.bar_label(container)
    
plt.show()
```

```python
ax = sns.catplot(
    data=labour_slack,
    kind='bar',
    x='TIME',
    y='Value',
    hue='SEX',
    col='GEO',
    height=6,
)

ax.set_axis_labels('', 'Labour slack (%)')
ax.fig.subplots_adjust(top=0.8) 
ax.fig.suptitle('Labour slack by sex in Poland and Spain');
```

**Выводы:**

- в целом избыток на рынке труда в Испании выше чем в Польше
- в Испании избыток среди женщин значительно выше, чем у мужчин. Также к 2021 году разрыв между мужчинами и женщинами - увеличился и уже составляет почти 10 процентов. В 2011 году этот разрыв был равен 8 процентам.
- в Польше избыток на рынке труда сократился за 10 лет почти в два раза как для женщин так и для мужчин
- также в Польше сократилась разница между мужчинами и женщинами. 3.4% в 2011 году, против 0.9% в 2021 году.
